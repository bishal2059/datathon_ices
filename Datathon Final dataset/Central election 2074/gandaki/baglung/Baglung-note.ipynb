```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import math
```

```python
df1 = pd.read_csv('1.csv')
df2 = pd.read_csv('2.csv')
```

```python
df1['EStatus'].fillna('Not Elected',inplace=True)
```

```python
random = np.random.randint(150,256)
df1['TotalVotesReceived'].fillna(random,inplace=True)
```

```python
re1 = df1['CandidateName'][0:4]
pt.rcParams['font.family'] = 'Utsaah'
x = np.array(re1)
re2 = df1['TotalVotesReceived'][0:4]
y = np.array(re2)
col =np.array(['red','green','blue','grey'])
pt.pie(y,labels=x,colors=col)
pt.title('Top 4 Candiadte Central Election Result for Baglung Area1')
pt.legend(title='Candidate Names',labels=y)
pt.show()
```

```python
df2['Gender'].fillna('पुरुष',inplace=True)
```

```python
df2['EStatus'].fillna('Not Elected',inplace=True)
```

```python
re1 = df2['CandidateName'][0:4]
pt.rcParams['font.family'] = 'Utsaah'
x = np.array(re1)
re2 = df2['TotalVotesReceived'][0:4]
y = np.array(re2)
col =np.array(['red','green','blue','grey'])
pt.pie(y,labels=x,colors=col)
pt.title('Top 4 Candiadte Central Election Result for Baglung Area2')
pt.legend(title='Candidate Names',labels=y)
pt.show()
```
