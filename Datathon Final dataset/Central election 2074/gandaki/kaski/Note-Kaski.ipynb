```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import math 
```

```python
df1 = pd.read_csv('1.csv')
df2 = pd.read_csv('2.csv')
df3 = pd.read_csv('3.csv')
```

```python
df1["StateID"].fillna(4.0,inplace=True)
df1['CenterConstID'].fillna(1.0,inplace=True)
df1['EStatus'].fillna('Not Elected',inplace=True)
```

```python
df2['EStatus'].fillna('Not Elected',inplace=True)
df2['SerialNo'].fillna(1325.0,inplace=True)
random = np.random.randint(108,130)
df2['TotalVotesReceived'].fillna(random,inplace=True)
```

```python
df3['EStatus'].fillna('Not Elected',inplace=True)
df3.loc[3,'Age']=df3['Age'].median()
df3['CenterConstID'].fillna(3.0,inplace=True)
```

```python
df1.to_csv('clean1.csv')
df2.to_csv('clean2.csv')
df3.to_csv('clean3.csv')
```

```python
re1 = df1['CandidateName'][0:4]
pt.rcParams['font.family'] = 'Utsaah'
x = np.array(re1)
re2 = df1['TotalVotesReceived'][0:4]
y = np.array(re2)
col =np.array(['red','green','blue','grey'])
pt.pie(y,labels=x,colors=col)
pt.title('Top 4 Candiadte Central Election Result for Kaski Area1')
pt.legend(title='Candidate Names',labels=y)
pt.show()
```
