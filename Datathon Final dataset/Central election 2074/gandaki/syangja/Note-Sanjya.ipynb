```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import math
```

```python
df1 = pd.read_csv('1.csv')
df2 = pd.read_csv('2.csv')
```

```python
df1
```

```python
for x in df1['CenterConstID']:
    if x != 1:
        y=1
        df1.loc[x,'CenterConstID'] = y
```

```python
df1
```

```python
df1['EStatus'].fillna('Not Elected',inplace=True)
```

```python
df1
```

```python
x = np.array(df1['CandidateName'])
y= np.array(df1['TotalVotesReceived'])
pt.rcParams['font.family'] = 'Utsaah'
z = np.array(['red','green','blue','grey','yellow','orange'])
pt.bar(x,y,color =z)
pt.legend(title='Winning Candidate',labels=y)
pt.title('Top 6 Candidate Central Election Result in Sanjya Area1')
pt.show()
```
